```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
source  = requests.get('https://www.imdb.com/list/ls025646548/')
```

```python
soup = BeautifulSoup(source.content)
```

```python
names = soup.find('h3',class_="lister-item-header")


names.text.replace("\n","")
```

```python
header = []
releaseyear=[]
```

```python
for i in soup.find_all('h3',class_="lister-item-header"):
    header.append(i.text.split("\n")[2])
    releaseyear.append(i.text.split("\n")[3])

print(releaseyear)
```

```python
rating = soup.find('span',class_="ipl-rating-star__rating")
```

```python
ratings = []
ratings_clean=[]
counter=0

for i in soup.find_all('span',class_="ipl-rating-star__rating"):
    ratings.append(i.text)

while(counter<len(ratings)):
    ratings_clean.append(ratings[counter])
    counter=counter+23
    
```

```python
import pandas as pd
df=pd.DataFrame({'Header': header[0:50],'Rating':ratings_clean, 'ReleaseYear':releaseyear[0:50]})

df
```
