```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
source = requests.get('https://www.cnbc.com/world/?region=world')
```

```python
soup = BeautifulSoup(source.content)
```

```python
header=[]
time=[]
link=[]

newsdata=soup.find('ul',class_='LatestNews-list').find_all('li')

for news in newsdata:
    header.append(news.find('div', class_='LatestNews-headlineWrapper').find('a').text)
    time.append(news.find('div', class_='LatestNews-headlineWrapper').find('time').text)
#     defined single variable for extracting link href
    linkall=news.find('div', class_='LatestNews-headlineWrapper').find('a', href=True)
    link.append(linkall['href'])
    
print(len(link))
print(*link, sep="\n")
```

```python
import pandas as pd
df = pd.DataFrame({'Header':header, 'Time':time, 'URL':link})
df
```
