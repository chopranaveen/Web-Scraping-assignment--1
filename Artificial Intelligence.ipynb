```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
source = requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
```

```python
soup = BeautifulSoup(source.content)
```

```python
paperTitle=[]
author=[]
publishedDate=[]
pageUrl=[]

rawdata=soup.find('ul', class_='sc-9zxyh7-0 cMKaMj').find_all('li')

for data in rawdata:
    paperTitle.append(data.find('a', class_='sc-5smygv-0 fIXTHm').find('h2').text)
    author.append(data.find('span', class_='sc-1w3fpd7-0 dnCnAO').text)
    publishedDate.append(data.find('span', class_='sc-1thf9ly-2 dvggWt').text)
    urlData=data.find('a', class_='sc-5smygv-0 fIXTHm', href=True)
    pageUrl.append(urlData['href'])
    
```

```python
import pandas as pd
df=pd.DataFrame({'Title':paperTitle, 'Author':author, 'Published Date':publishedDate, 'URL':pageUrl})
df
```
