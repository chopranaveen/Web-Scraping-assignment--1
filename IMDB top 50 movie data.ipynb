```python
from bs4 import BeautifulSoup
import requests
```

```python
 source= requests.get('https://www.imdb.com/chart/top/')
```

```python
soup= BeautifulSoup(source.text,'html.parser') 
```

```python
movies = soup.find('tbody', class_="lister-list").find_all('tr')
name=[]
rank=[]
year=[]
rating=[]

for movie in movies:
    
    name.append(movie.find('td',class_="titleColumn").find('a').text)
    
    rank.append(movie.find('td',class_="titleColumn").get_text(strip=True).split('.')[0])
    
    year.append(movie.find('td',class_="titleColumn").span.text.strip('()'))
    
    rating.append(movie.find('td',class_="ratingColumn").strong.text)
    
    
    
```

```python
import pandas as pd
df = pd.DataFrame({'Name':name[0:50], 'Rank':rank[0:50], 'Year':year[0:50], 'Rating':rating[0:50]})
df
```
